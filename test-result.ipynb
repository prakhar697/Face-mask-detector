```python
import tensorflow as tf
```

```python
loaded_model = tf.keras.models.load_model('model.h5')
```

```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
folder = 'Dataset/with_mask'
```

```python
img = []
for filename in os.listdir(folder):
    img_array = cv2.imread(os.path.join(folder,filename))
    img.append(img_array)
```

```python
img
```

```python
plt.imshow(img[4])
```

```python
#img.shape
```

```python
IMG_SIZE = 224
new_array = []

for i in img:
    new_array.append(cv2.resize(i, (IMG_SIZE,IMG_SIZE)))
```

```python
new_array
```

```python
x = []

for i in new_array:
    
    x.append(np.array(i).reshape(-1,IMG_SIZE,IMG_SIZE,3)) 
#x.shape
```

```python
result = []

for i in x:
    
    result.append(loaded_model.predict(i))
```
